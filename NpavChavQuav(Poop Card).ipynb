```python
import random
import time
from collections import Counter
```

```python
suits = ['♠', '♥', '♦', '♣']
cards = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
deck = []
poopcard = []
hand1 = []
hand2 = []
for card in cards:
    for suit in suits:
        deck.append(f"{card}{suit}")   

random.shuffle(deck)
random.shuffle(deck)
print("Shuffling Deck...")
time.sleep(2)
poopcard.append(deck[30])
print("Choosing the poop card...")
time.sleep(2)
del (deck[30])
print("Poop Card Chosen! Dealing Cards!")
for i, item in enumerate(deck):
    if i % 2 == 0:
        hand1.append(item)
    else:
        hand2.append(item)
time.sleep(2)
print("\nCards dealt!")
time.sleep(0.5)
print(f"Player 1 Hand: {hand1}")
print(" ")
print(f"Player 2 Hand: ???")
time.sleep(0.5)
print("Removing pairs from each hand...")
time.sleep(1)
print("\n")
#---------------------------------------------------hand 1------------------------------------
prefix_count_hand1 = Counter(card[:-1] for card in hand1)
result_hand1 = []
removed_count_hand1 = {}

for card in hand1:
    prefix_hand1 = card[:-1]
    count_hand1 = prefix_count_hand1[prefix_hand1]

    if prefix_hand1 not in removed_count_hand1:
        removed_count_hand1[prefix_hand1] = 0

    pairs_to_remove_hand1 = (count_hand1 // 2) * 2

    if removed_count_hand1[prefix_hand1] < pairs_to_remove_hand1:
        removed_count_hand1[prefix_hand1] += 1
    else:
        result_hand1.append(card)
#---------------------------------------------------hand 2------------------------------------
prefix_count_hand2 = Counter(card[:-1] for card in hand2)
result_hand2 = []
removed_count_hand2 = {}

for card in hand2:
    prefix_hand2 = card[:-1]
    count_hand2 = prefix_count_hand2[prefix_hand2]

    if prefix_hand2 not in removed_count_hand2:
        removed_count_hand2[prefix_hand2] = 0

    pairs_to_remove_hand2 = (count_hand2 // 2) * 2
    
    if removed_count_hand2[prefix_hand2] < pairs_to_remove_hand2:
        removed_count_hand2[prefix_hand2] += 1
    else:
        result_hand2.append(card)
#----------------------------------------------------------------------------------------
print("Player 1 Remaining cards:", result_hand1)
print("Cards:", len(result_hand1))
print("\nPlayer 2 Remaining cards: ???")
print("Cards:", len(result_hand2))
#----------------------------------------------------------------------------------------
while len(result_hand1) > 0 and len(result_hand2) > 0:
#player 1
    print("\n" + "="*50)
    print("PLAYER 1'S TURN")
    print("="*50)
    print(f"\nYour hand: {result_hand1}")
    print(f"You have {len(result_hand1)} cards")
    print(f"\nPlayer 2 has {len(result_hand2)} cards")
    
    print("\nChoose a card from Player 2:")
    for i in range(len(result_hand2)):
        print(f"  {i+1}. Card {i+1}")
    
    while True:
        try:
            choice = int(input(f"\nEnter a number (1-{len(result_hand2)}): "))
            if 1 <= choice <= len(result_hand2):
                break
            else:
                print(f"Please enter a number between 1 and {len(result_hand2)}")
        except ValueError:
            print("Please enter a valid number")
    
    chosen_card = result_hand2.pop(choice - 1)
    print(f"\nYou drew: {chosen_card}")
    result_hand1.append(chosen_card)

    hand_size_before = len(result_hand1)
    
    prefix_count_hand1 = Counter(card[:-1] for card in result_hand1)
    temp_hand1 = []
    removed_count_hand1 = {}
    for card in result_hand1:
        prefix_hand1 = card[:-1]
        count_hand1 = prefix_count_hand1[prefix_hand1]
        if prefix_hand1 not in removed_count_hand1:
            removed_count_hand1[prefix_hand1] = 0
        pairs_to_remove_hand1 = (count_hand1 // 2) * 2
        if removed_count_hand1[prefix_hand1] < pairs_to_remove_hand1:
            removed_count_hand1[prefix_hand1] += 1
        else:
            temp_hand1.append(card)
    result_hand1 = temp_hand1
 
    if len(result_hand1) < hand_size_before:
        print(f"\nYOU GOT A PAIR!")
    else:
        print(f"\nNo pairs formed.")
    
    print(f"Your hand: {result_hand1}")
    print(f"You now have {len(result_hand1)} cards")
    
    time.sleep(2)
    if len(result_hand1) == 0:
        print("\n" + "="*50)
        print("PLAYER 1 WINS!")
        print("="*50)
        print(f"The poop card was: {poopcard[0]}")
        break
    if len(result_hand2) == 0:
        print("\n" + "="*50)
        print("PLAYER 1 LOSES!")
        print(f"Player 1 is stuck with: {result_hand1}")
        print(f"The poop card was: {poopcard[0]}")
        print("="*50)
        break
    print("\n" + "="*50)
    print("PLAYER 2'S TURN (Computer)")
    print("="*50)
    time.sleep(1)
    computer_choice = random.randint(0, len(result_hand1) - 1)
    chosen_card_by_computer = result_hand1.pop(computer_choice)
    print("\nPlayer 2 drew a card from your hand")
    time.sleep(1)
    print(f"Player 2 took: {chosen_card_by_computer}")
    result_hand2.append(chosen_card_by_computer)
    prefix_count_hand2 = Counter(card[:-1] for card in result_hand2)
    temp_hand2 = []
    removed_count_hand2 = {}
    pairs_formed = False  
    
    for card in result_hand2:
        prefix_hand2 = card[:-1]
        count_hand2 = prefix_count_hand2[prefix_hand2]
        if prefix_hand2 not in removed_count_hand2:
            removed_count_hand2[prefix_hand2] = 0
        pairs_to_remove_hand2 = (count_hand2 // 2) * 2
        if removed_count_hand2[prefix_hand2] < pairs_to_remove_hand2:
            removed_count_hand2[prefix_hand2] += 1
            pairs_formed = True  
        else:
            temp_hand2.append(card)
    result_hand2 = temp_hand2
    
    if pairs_formed:
        print("\nPLAYER 2 GOT A PAIR!")
    else:
        print("\nNo pairs formed.")
    
    print(f"Player 2 now has {len(result_hand2)} cards")
    
    time.sleep(2)
    if len(result_hand2) == 0:
        print("\n" + "="*50)
        print("PLAYER 2 WINS!")
        print("="*50)
        print(f"The poop card was: {poopcard}")
        break
    if len(result_hand1) == 0:
        print("\n" + "="*50)
        print("PLAYER 1 WINS!")
        print(f"Player 2 is stuck with: {result_hand2}")
        print(f"The poop card was: {poopcard}")
        print("="*50)
        break
```
