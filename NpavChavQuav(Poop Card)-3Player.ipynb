```python
import random
import time
from collections import Counter
```

```python
# Run this cell to start the game of Npav Chav Quav
suits = ['♠', '♥', '♦', '♣']
cards = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
deck = []
poopcard = []
hand1 = []
hand2 = []
hand3 = []

for card in cards:
    for suit in suits:
        deck.append(f"{card}{suit}")   

random.shuffle(deck)
random.shuffle(deck)
print("Shuffling Deck...")
time.sleep(1)

poopcard.append(deck[30])
print("Choosing the poop card...")
time.sleep(1)
del (deck[30])
print("Poop Card Chosen! Dealing Cards to 3 Players!")

for i, item in enumerate(deck):
    if i % 3 == 0:
        hand1.append(item)
    elif i % 3 == 1:
        hand2.append(item)
    else:
        hand3.append(item)

time.sleep(1)
print("\nCards dealt!")
time.sleep(0.5)
print(f"Player 1 Hand: {len(hand1)} Cards: {hand1}")
print(" ")
print(f"Player 2 Hand: {len(hand2)} Cards")
print(" ")
print(f"Player 3 Hand: {len(hand3)} Cards")
time.sleep(0.5)

# -------------------- Remove pair function --------------------
def remove_pairs(hand):
    prefix_count = Counter(card[:-1] for card in hand)
    result_hand = []
    removed_count = {}

    for card in hand:
        prefix = card[:-1]
        count = prefix_count[prefix]
        if prefix not in removed_count:
            removed_count[prefix] = 0
        pairs_to_remove = (count // 2) * 2
        if removed_count[prefix] < pairs_to_remove:
            removed_count[prefix] += 1
        else:
            result_hand.append(card)
    return result_hand

# -------------------- Remove pair function --------------------

print("\nRemoving pairs from each hand...")
result_hand1 = remove_pairs(hand1)
result_hand2 = remove_pairs(hand2)
result_hand3 = remove_pairs(hand3)
time.sleep(1)

print(f"\nPlayer 1 (You) Cards: {len(result_hand1)}")
print(f"Player 2 (Comp) Cards: {len(result_hand2)}")
print(f"Player 3 (Comp) Cards: {len(result_hand3)}")

# -------------------- GAME LOOP --------------------
while (len(result_hand1) + len(result_hand2) + len(result_hand3)) > 1:
    
# ------------------ PLAYER 1 ------------------
    if len(result_hand1) > 0:
        print("\n" + "-"*50)
        print("PLAYER 1'S TURN (You)")
        print("-"*50)
        print(f"Your hand: {result_hand1}")

        valid_targets = []
        if len(result_hand2) > 0: valid_targets.append(2)
        if len(result_hand3) > 0: valid_targets.append(3)
        
        if len(valid_targets) == 0:
             print("Everyone else is out! You are stuck with the last card!")
             break

        target_hand = None
        target_name = ""
        
        if len(valid_targets) == 1:
            p_choice = valid_targets[0]
            print(f"Only Player {p_choice} has cards left.")
        else:
            while True:
                try:
                    p_choice = int(input(f"Who do you want to steal from? (Enter {valid_targets}): "))
                    if p_choice in valid_targets:
                        break
                    else:
                        print(f"Invalid choice. Choose {valid_targets}")
                except ValueError:
                    print("Invalid input.")

        if p_choice == 2:
            target_hand = result_hand2
            target_name = "Player 2"
        else:
            target_hand = result_hand3
            target_name = "Player 3"

        print(f"\nChoose a card from {target_name}:")
        for i in range(len(target_hand)):
            print(f"  {i+1}. Card {i+1}")
        
        while True:
            try:
                c_choice = int(input(f"Enter a number (1-{len(target_hand)}): "))
                if 1 <= c_choice <= len(target_hand):
                    break
                else:
                    print(f"Please enter a number between 1 and {len(target_hand)}")
            except ValueError:
                print("Please enter a valid number")
        
        chosen_card = target_hand.pop(c_choice - 1)
        print(f"\nYou drew: {chosen_card}")
        result_hand1.append(chosen_card)

        hand_size_before = len(result_hand1)
        result_hand1 = remove_pairs(result_hand1)
     
        if len(result_hand1) < hand_size_before:
            print(f"YOU GOT A PAIR!")
        else:
            print(f"No pairs formed.")
        
        print(f"Your hand now: {result_hand1}")
        if len(result_hand1) == 0:
            print("\n*** YOU HAVE 0 CARDS! YOU'RE OUT! ***")
            
        time.sleep(2)

    if (len(result_hand1) + len(result_hand2) + len(result_hand3)) <= 1: break


# ------------------ PLAYER 2 (CPU 1) ------------------
    if len(result_hand2) > 0:
        print("\n" + "-"*50)
        print("PLAYER 2'S TURN (CPU 1)")
        print("-"*50)

        valid_targets = []
        if len(result_hand1) > 0: valid_targets.append(1)
        if len(result_hand3) > 0: valid_targets.append(3)
        
        if len(valid_targets) > 0:
            target_choice = random.choice(valid_targets)
            
            if target_choice == 1:
                print("Player 2 decides to steal from YOU!")
                steal_source = result_hand1
                source_name = "Player 1"
            else:
                print("Player 2 decides to steal from Player 3!")
                steal_source = result_hand3
                source_name = "Player 3"
                
            time.sleep(1)

            rand_idx = random.randint(0, len(steal_source) - 1)
            stolen_card = steal_source.pop(rand_idx)
            
            if target_choice == 1:
                print(f"Player 2 took: {stolen_card}")
            else:
                print("Player 2 took a card from Player 3.")

            result_hand2.append(stolen_card)

            hand_size_before_p2 = len(result_hand2)
            result_hand2 = remove_pairs(result_hand2)
            
            if len(result_hand2) < hand_size_before_p2:
                print("PLAYER 2 GOT A PAIR!")
            else:
                print("No pairs formed.")
                
            print(f"Player 2 has *{len(result_hand2)}* cards")
            if len(result_hand2) == 0:
                print("\n*** PLAYER 2 IS OUT! ***")
        else:
            print("Player 2 has no one to steal from!")

        time.sleep(2)

    if (len(result_hand1) + len(result_hand2) + len(result_hand3)) <= 1: break


# ------------------ PLAYER 3 (CPU 2) ------------------
    if len(result_hand3) > 0:
        print("\n" + "-"*50)
        print("PLAYER 3'S TURN (CPU 2)")
        print("-"*50)
        
        valid_targets = []
        if len(result_hand1) > 0: valid_targets.append(1)
        if len(result_hand2) > 0: valid_targets.append(2)
        
        if len(valid_targets) > 0:
            target_choice = random.choice(valid_targets)
            
            if target_choice == 1:
                print("Player 3 decides to steal from YOU!")
                steal_source = result_hand1
                source_name = "Player 1"
            else:
                print("Player 3 decides to steal from Player 2!")
                steal_source = result_hand2
                source_name = "Player 2"
                
            time.sleep(1)
            
            rand_idx = random.randint(0, len(steal_source) - 1)
            stolen_card = steal_source.pop(rand_idx)
            
            if target_choice == 1:
                print(f"Player 3 took: {stolen_card}")
            else:
                print("Player 3 took a card from Player 2.")

            result_hand3.append(stolen_card)
            
            hand_size_before_p3 = len(result_hand3)
            result_hand3 = remove_pairs(result_hand3)
            
            if len(result_hand3) < hand_size_before_p3:
                print("PLAYER 3 GOT A PAIR!")
            else:
                print("No pairs formed.")
                
            print(f"Player 3 has *{len(result_hand3)}* cards")
            if len(result_hand3) == 0:
                print("\n*** PLAYER 3 IS OUT! ***")
        else:
            print("Player 3 has no one to steal from!")
            
        time.sleep(2)

# -------------------- GAME OVER --------------------
print("\n" + "-"*50)
print("GAME OVER")
print("-"*50)


if len(result_hand1) > 0:
    print(f"\nPLAYER 1 (YOU) LOST!")
    print(f"You have: {result_hand1[0]}")
    print(f"The poop card was: {poopcard[0]}")
elif len(result_hand2) > 0:
    print(f"\nPLAYER 2 LOST!")
    print(f"Player 2 is stuck with: {result_hand2[0]}")
    print(f"The poop card was: {poopcard[0]}")
elif len(result_hand3) > 0:
    print(f"Player 3 is stuck with: {result_hand3[0]}")
    print(f"The poop card was: {poopcard[0]}")
```
